## SP500

```python
Y_day = pd.read_excel('S&P 500.xls')
Y_hour = pd.read_excel('S&P 500_new.xls')
Y_day
```

```python
Y_day['Amplitude'] = Y_day['Close'].shift(-1) > Y_day['Close'] * 1

Y_hour['Amplitude'] = Y_hour['Close'].shift(-1) > Y_hour['Close'] * 1
```

```python
#Y_day
Y_hour
```

```python
Y_day['IS_UP'] = Y_day['Amplitude'] > 0
Y_hour['IS_UP'] = Y_hour['Amplitude'] > 0
```

```python
print(Y_day['IS_UP'].value_counts())
print(Y_hour['IS_UP'].value_counts())
```
