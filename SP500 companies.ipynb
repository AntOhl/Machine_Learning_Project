```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
companies = pd.read_excel('SP500_companies.xlsx')
```

```python
companies
```

```python
companies_sorted = companies.sort_values(by=['Market Cap ($M)'])
```

```python
plt.style.use('dark_background')
companies_sorted['Market Cap ($M)'].plot.bar(color = 'orange', figsize = (20, 7))
plt.title('S&P500 Market Cap Distribution', fontsize = 30, fontweight = 20)
plt.xlabel('Name of Companies')
plt.ylabel('Market Cap')
plt.show()
```

```python
is_large = companies['Market Cap ($M)'] > 180000

```

```python
companies_large = companies[is_large]
companies_large
```

```python
companies_large = companies_large.Name.apply(lambda x : x.replace(' Inc.',""))
companies_large
```

```python
companies_large = companies_large.apply(lambda x : x.replace(' Corp.',""))
companies_large
```

```python
companies_large = companies_large.apply(lambda x : x.replace(' & Co',''))
companies_large
```

```python
companies_large = companies_large.apply(lambda x : x.replace(' Co.',''))
companies_large
```

```python
companies_large = companies_large.apply(lambda x : x.replace('The ',''))
companies_large
```

```python
companies_large = companies_large.apply(lambda x : x.replace(',',''))
companies_large
```

```python
companies_large = companies_large.apply(lambda x : x.replace('.com',''))
companies_large
```

```python
companies_large = companies_large.apply(lambda x : x.replace('.',''))
companies_large
```

```python
companies_large = companies_large.apply(lambda x : x.replace(' Group',''))
companies_large
```

```python
companies_large = companies_large.apply(lambda x : x.replace(' Systems',''))
companies_large
```

```python
companies_large = companies_large.apply(lambda x : x.replace(' Mobil',''))
companies_large
```
