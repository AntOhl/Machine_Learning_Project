```python
import spacy
from spacy.tokens import DocBin
import json
import pandas as pd
import pickle
import numpy as np
```

```python
nlp = spacy.load('en_core_web_md')
```

```python
load_pickletweet = pickle.load(open('tweets_new.p','rb'))
doc_bin = DocBin().from_bytes(load_pickletweet)
tl = list(doc_bin.get_docs(nlp.vocab))
#tl
```

```python
# Get lemmas 

def get_lemma(tl, n_tweets):
    for tweet in tl[:n_tweets]:
        for token in tweet:
            print("{attribute}{output} | ".format(attribute="TEXT > ", output=token.text),
                  "{attribute}{output} | ".format(attribute="LEMMA > ", output=token.lemma_))
```

```python
get_lemma(tl,10)
```

```python
# Get all attribute 

def get_signature(tl,n_tweets):
     for tweet in tl[:n_tweets]:
        for token in tweet:
            print("{attribute}{output} | ".format(attribute="TOKEN > ", output=token.text),
                  "{attribute}{output} | ".format(attribute="LEMMA > ", output=token.lemma_),
                  "{attribute}{output} | ".format(attribute="NATURE > ", output=token.pos_),
                  "{attribute}{output} | ".format(attribute="FUNCTION > ", output=token.dep_),
                  "{attribute}{output} | ".format(attribute="ENTITY > ", output=token.ent_type_))
```

```python
get_signature(tl,10)
```

```python
# Get all named entities and their label

def get_namedentities(tl,n_tweets):
    NAMED = []
    for tweet in tl[:n_tweets]:
        for ent in tweet.ents:
            NAMED.append(("{attribute}{output} | ".format(attribute="TOKEN > ", output=ent.text),
                              "{attribute}{output} | ".format(attribute="NATURE > ", output=ent.label_),
                              "{attribute}{output} | ".format(attribute="INDEX > ", output=ent.start_char)))
    return NAMED
```

```python
get_namedentities(tl,10)
```

```python
def get_political(tl,n_tweets):
    POLITICAL = []
    for tweet in tl[:n_tweets]:
        for ent in tweet.ents:
            if ent.label_ in ['NORP','GPE']:
                POLITICAL.append(("{attribute}{output} | ".format(attribute="TOKEN > ", output=ent.text),
                  "{attribute}{output} | ".format(attribute="NATURE > ", output=ent.label_),
                  "{attribute}{output} | ".format(attribute="INDEX > ", output=ent.start_char)))
    return POLITICAL
```

```python
political = get_political(tl,1000)
political
```

```python
def get_organization(tl,n_tweets):
    ORG = []
    for tweet in tl[:n_tweets]:
        for ent in tweet.ents:
            if ent.label_ in ['ORG']:
                if ent.text not in ['RT']:
                    ORG.append(("{attribute}{output} | ".format(attribute="TOKEN > ", output=ent.text),
                  "{attribute}{output} | ".format(attribute="NATURE > ", output=ent.label_),
                  "{attribute}{output} | ".format(attribute="INDEX > ", output=ent.start_char)))
    return ORG
```

```python
ORG = get_organization(tl,1000)
ORG
```

```python
def get_event(tl,n_tweets):
    EVENT = []
    for tweet in tl[:n_tweets]:
        for ent in tweet.ents:
            if ent.label_ in ['EVENT']:
                EVENT.append(("{attribute}{output} | ".format(attribute="TOKEN > ", output=ent.text),
                  "{attribute}{output} | ".format(attribute="NATURE > ", output=ent.label_),
                  "{attribute}{output} | ".format(attribute="INDEX > ", output=ent.start_char)))
    return EVENT
```

```python
event = get_event(tl,1000)
event
```
